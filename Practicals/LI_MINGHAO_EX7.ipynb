```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from NEMtropy import UndirectedGraph, DirectedGraph
from NEMtropy.network_functions import build_adjacency_from_edgelist
```

```python
cormat1 = np.load('cormat_1h.npy')
cormat2 = np.load('cormat_1m.npy')
cormat3 = np.load('cormat_gaussian_1h.npy')
cormat4 = np.load('cormat_gaussian_1m.npy')
cormat5 = np.load('cormat_onefactor_1h.npy')
cormat6 = np.load('cormat_onefactor_1m.npy')
```

```python
def MST(m):
    d = np.sqrt(2*(1-m)) 
    g = nx.from_numpy_array(d)
    t = nx.minimum_spanning_tree(g)
    return t
```

```python
def deg_cal(t):
    dict1 = {}
    f = open("stocknames.txt","r") 
    lines = f.readlines()
    i = 0
    for line in lines:
        dict1.update({i:str(line).replace('\n', '')})
        i=i+1
    nx.set_node_attributes(t, dict1, 'name')
    dict2 = nx.get_node_attributes(t, 'name')
    degree = t.degree()
    ndegrees = np.array([ degree[i] for i in t.nodes() ])
    
    list2 = dict2.values()
    dictionary = dict(zip(list2, ndegrees))
    return dictionary,dict1
```

```python
t1 = MST(cormat1)
t2 = MST(cormat2)
t3 = MST(cormat3)
t4 = MST(cormat4)
t5 = MST(cormat5)
t6 = MST(cormat6)
```
