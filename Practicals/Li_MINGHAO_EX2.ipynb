```python
#plot the average degree of the nearest neighbours knn
def ave_nei_deg(g):
    kn = nx.average_degree_connectivity(g)
    x = list(kn.keys())
    y = list(kn.values())
    plt.scatter(x,y)
    plt.xlabel('degree')
    plt.ylabel('degree knn')
    plt.suptitle(g)
    plt.show()
```

# A02.2 Clustering and randomisation

```python
#plot the degree distribution
def deg_dis(g):
    nodes = [i for i in g.nodes() ]
    degrees = [g.degree(i) for i in nodes]
    ave_deg = sum(degrees)/len(nodes)
    my_bins = np.logspace(np.log10(min(degrees)),np.log10(max(degrees) +1), 30 )
    _ = plt.hist(degrees, bins = my_bins,density = True)
    plt.axvline(ave_deg,color='r',linestyle='--',label='average degree<k>')
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('degree')
    plt.ylabel('density')
```

```python
#compute the clustering coefficient in the graphs
#scatter plot of c and crnd
def clu_coe(g):
    c = nx.clustering(g)
    ci = list(c.values())
    i = list(c.keys())
    grnd = nx.algorithms.smallworld.random_reference(g,niter =20, connectivity=False)
    crnd = nx.clustering(grnd)
    crndi = list(crnd.values())
    i2 = list(crnd.keys())
    plt.subplot(1, 2, 1)
    plt.scatter(i,ci)
    my_x_ticks = np.arange(0, int(i[-1]),int(i[-1])/5)
    plt.xticks(my_x_ticks)
    plt.xlabel('node')
    plt.ylabel('clustering')
    plt.title("real network")
    plt.subplot(1, 2, 2)
    plt.scatter(i2,crndi)
    my_x_ticks = np.arange(0, int(i2[-1]),int(i2[-1])/5)
    plt.xticks(my_x_ticks)
    plt.xlabel('node')
    plt.title("randomised network")
    return c,crnd
```
