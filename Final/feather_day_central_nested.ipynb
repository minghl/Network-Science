```python
def ajc_sort(g):
    adj = nx.to_numpy_array(g)
    arr = []
    for i in range(len(adj)):
        x = 0
        for j in range(len(adj)):
            if adj[i][j]!=0:
                x=x+1
        arr.append(x)
    d = np.insert(adj, len(adj), values=arr, axis=1)
    dd = d[np.lexsort(-d.T)]
    dataset = np.delete(dd, -1, axis=1)
    da = dataset.T
    arr2 = []
    for i in range(len(da)):
        x = 0
        for j in range(len(da)):
            if da[i][j]!=0:
                x=x+1
        arr2.append(x)
    d2 = np.insert(da, len(da), values=arr2, axis=1)
    dd2 = d2[np.lexsort(-d2.T)]
    dataset2 = np.delete(dd2, -1, axis=1)
    dataset3= dataset2.T
    
    return dataset3
```

```python
def nestedness(g):
    adj_kar = nx.to_numpy_array(g)
    adj = adj_kar[[not np.all(adj_kar[i] == 0) for i in range(adj_kar.shape[0])], :]
    row_idxs1 = [ set(np.where(i == 1)[0]) for i in adj ]
    row_idxs1 = [value for value in row_idxs1 if value != set()]
    row_pairs1 = list(it.combinations(row_idxs1, 2))
    Np_row1 = [ 0 if len(l) >= len(u) else 100*len(u & l) / len(l) for u, l in row_pairs1 ]
    a = adj_kar.transpose()
    a = a[[not np.all(a[i] == 0) for i in range(a.shape[0])], :]
    col_idxs1 = [ set(np.where(i == 1)[0]) for i in a ]
    col_idxs1 = [value for value in col_idxs1 if value != set()]
    col_pairs1 = list(it.combinations(col_idxs1, 2))
    Np_col1 = [ 0 if len(l) >= len(u) else 100*len(u & l) / len(l) for u, l in col_pairs1 ]
    numerator = sum(Np_row1)+sum(Np_col1)
    N = (len(Np_row1)*(len(Np_row1)-1))/2
    M = (len(Np_col1)*(len(Np_col1)-1))/2
    denominator = N+M
    nestedness = numerator/denominator

    return nestedness
```

```python
def col(g_list):
    nes_list = []
    for g in g_list:
        nes_list.append(nestedness(g))
    p_p = sp.pearsonr(nes_list,pcl_list)[0]
    k_p = sp.pearsonr(nes_list,kcl_list)[0]
    s_p = sp.pearsonr(nes_list,scl_list)[0]
    p_k = sp.kendalltau(nes_list,pcl_list)[0]
    k_k = sp.kendalltau(nes_list,kcl_list)[0]
    s_k = sp.kendalltau(nes_list,scl_list)[0]
    p_s = sp.spearmanr(nes_list,pcl_list)[0]
    k_s = sp.spearmanr(nes_list,kcl_list)[0]
    s_s = sp.spearmanr(nes_list,scl_list)[0]
    cor_list = [p_p,k_p,s_p,p_k,k_k,s_k,p_s,k_s,s_s]
    b = [round(i,3) for i in cor_list]
    cor_name = ["p_p","k_p","s_p","p_k","k_k","s_k","p_s","k_s","s_s"]
    plt.figure(figsize=(8, 5))
    plt.xlabel("Different types of correlation coefficient")
    plt.ylabel("Correlation coefficient value")
    plt.title("Relation between nestedness and centrality metrics' correlation")
    x = plt.bar(cor_name, b, width=1, edgecolor="white", linewidth=0.7)
    plt.bar_label(x, padding=3)
    return nes_list,cor_list
```
